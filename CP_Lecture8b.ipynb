#Simple pendulum

Use Python Module scipy.integrate.ode to study the differential equations of a simple pendulum

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
```

Consider the differential equation for a real pendulum below:

```
# define function describing the differential equation
def f(t, z, alpha):
  theta, v = z
  dfdt = [v, -np.sin(theta)-alpha*v]
  return dfdt
```

Inspect how scipy.integrate.ode can be used to solve the differential equation and plot the solutions

```
# solve the ODE
solver = ode(f)
solver.set_integrator('dopri5')
alpha = 0.0
solver.set_f_params(alpha)

# initial conditions
z0 = [0,0.1]
t0 = 0; t1 = 100
N = 2750
solver.set_initial_value(z0, t0)
t = np.linspace(t0, t1, N)
sol = np.empty((N, 2))
sol[0] = z0

k=1
while solver.successful() and solver.t < t1:
  solver.integrate(t[k])
  sol[k] = solver.y
  k += 1

plt.plot(t, sol[:,0], label = 'theta',linewidth=3)
plt.plot(t, sol[:,1], label = 'velocity')
plt.legend(loc='best')
plt.xlabel('time')
plt.ylabel('signal')
plt.xlim(0,20)
plt.show()
```
